```python
import pandas
```

```python
praha = pandas.read_csv ("zam_praha.txt")
liberec = pandas.read_csv("zam_liberec.txt")
plzen = pandas.read_csv("zam_plzeň.txt")
print(praha,liberec, plzen)
```

```python
praha["mesto"] = "Praha"
print(praha)
```

```python
liberec["mesto"] = "Liberec"
print(liberec)
```

```python
plzen["mesto"] = "Plzen"
print(plzen)
```

```python
zamestnanci = pandas.concat([praha, liberec, plzen])
print(zamestnanci)
```

```python
platy = pandas.read_csv ("platy_2021_02.txt")
```

```python
zamestnanec_plat = pandas.merge(zamestnanci, platy)
zamestnanec_plat
```

priemerný plat


```python
priemer_plat = zamestnanec_plat.groupby("mesto")["plat"].mean()
print(priemer_plat)
```
